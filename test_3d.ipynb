```sage
# Initial input
U0 = Polyhedron(ieqs=[[0, 1, 0],[0,0,1],[2,-1,-2]])
U0.plot()
```

```sage
def box(n, ep):
    zeros = [0]*(n+1)
    zeros[0] = ep
    mat = []
    for i in range(2*n):
        mat.append(zeros*1)
    # print(mat)

    for i in range(n):
        mat[i][i+1] = 1
    j = 1
    for i in range(n,2*n):
        mat[i][j] = -1
        j += 1

    Box = Polyhedron(ieqs=mat)
    return Box

Box = box(3,2)
Box.plot()
```
