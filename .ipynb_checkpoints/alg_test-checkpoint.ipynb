```python
import numpy as np
x0, u0 = vector([2,3,45]), 0.4
n, ep = 10, 3
ep_s = {ep, -ep}
ep_vers = [(i,j,k) for i in ep_s for j in ep_s for k in ep_s]

ep_vers = Polyhedron(vertices = ep_vers)

ep_vers.plot()
```

```python
xc, uc = x0, u0
theta = xc[2]
A = Matrix([[0,0,3*np.cos(theta)/theta],[0,0,3*np.sin(theta)/theta],[0,0,0]])
B = vector([0,0,1])
K = 1
```

```python
for i in range(n):
    xc = A*xc + B*uc
    print(xc.vertices())
    plot(xc)
    x1 = xc + ep_vers
    uc = x1*K
```
