```sage
import ppl
x = ppl.Variable(0)
y = ppl.Variable(1)
cs = ppl.Constraint_System()
cs.insert(x >= 0)
cs.insert(x <= 1)
cs.insert(y >= 0)
cs.insert(y <= 1)
poly = ppl.C_Polyhedron(cs)
poly
```

```sage
x1 = ppl.Variable(0)
y1 = ppl.Variable(1)
cs = ppl.Constraint_System()
cs.insert(x1 <= 0)
cs.insert(x1 >= -1)
cs.insert(y1 <= 0)
cs.insert(y1 >= -1)
poly2 = ppl.C_Polyhedron(cs)
poly2
```
